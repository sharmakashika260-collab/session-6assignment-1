```python
food_prices = {"Burger": 75, "Pizza": 350, "Pasta": 150, "Fries": 100, "Biryani": 220}

print(food_prices)

for item in food_prices:
    if food_prices[item] > 200:
        print(item, food_prices[item])
```

```python
flipkart_users = {"asmita", "neha", "reena", "jannat", "kiran"}
myntra_users = {"neha", "kiran", "sneha", "asmita", "rahul"}

users = flipkart_users.intersection(myntra_users)

print(users)
```
